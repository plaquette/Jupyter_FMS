```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0,20/360 * np.pi,0.0001)
y1 = np.sin(1/0.00227*x)
y2 = np.sin(1/0.015*x)

#[Sin[1/0.002*x], {x, 0, 1/360 \[Pi]}]

plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['sin(T=0.0027)', 'sin(T=0.015)'])
plt.show()
```
