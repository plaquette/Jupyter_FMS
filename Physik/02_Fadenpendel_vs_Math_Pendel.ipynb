Federpendel vs. Math. Pendel


```python
import numpy as np
from numpy import sin, cos
from scipy.integrate import odeint
from matplotlib import pyplot as plt
%matplotlib inline


# define the equations
def equations(y0, t):
	theta, x = y0
	f = [x, -(g/l) * sin(theta)]
	return f

def plot_results(time, theta1, theta2):
	plt.plot(time, theta1[:,0])
	plt.plot(time, theta2)

	s = '(Anfangswinkel = ' + str(initial_angle) + ' grad)'
	plt.title('Fadenpendel vs. Math. Pendel: ' + s)
	plt.xlabel('time (s)')
	plt.ylabel('Winkel (rad)')
	plt.grid(True)
	plt.legend(['Fadenpendel', 'Math. Pendel'], loc='lower right')
	plt.show()

# parameter
g = 9.81
l = 1.0

time = np.arange(0, 20.0, 0.025)

# Startbedingungen
initial_angle = 65.0
theta0 = np.radians(initial_angle)
x0 = np.radians(0.0)

# Diffgln. fr Fadenpendel
theta1 = odeint(equations, [theta0, x0],  time)

# Math. Pendel Lsung
w = np.sqrt(g/l)
theta2 = [theta0 * cos(w*t) for t in time]

# plot
plot_results(time, theta1, theta2)

#plt.show()

```

```python
x = np.arange(0,50/360 * np.pi,0.0001)
y1 = np.sin(x)
y2 = x

#[Sin[1/0.002*x], {x, 0, 1/360 \[Pi]}]

plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['sin(x)', 'x'])
plt.show()
```

Aus der Lnge des Pendels und der Schwingdauer des Federpendels (fr kleine Winkel) kann also die Fallbeschleunigung g bestimmt werden:

```python
T = 11.48/6
l = 0.9
def fallbeschleunigung(T, l):
    g  = (4 * np.pi**2 * l) / (T**2)
    return  g

print("die Fallbeschleunigung betrgt " + str(fallbeschleunigung(T,l)) + "m/s^2")
```
