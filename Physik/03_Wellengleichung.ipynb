Bsp. Wellengleichung


```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
%matplotlib inline

ymax = 0.02
TT = 8
lam = 0.2

x,t = np.meshgrid(np.linspace(0, 0.6), np.linspace(0, 50))


y = ymax * np.sin(2*np.pi * (t/TT - x/lam))

ax = plt.figure().gca(projection='3d')
ax.plot_surface(x, t, y, cmap=cm.coolwarm, linewidth=0, antialiased=False)

#ax.plot_wireframe(x,t,y)
#[Sin[1/0.002*x], {x, 0, 1/360 \[Pi]}]

plt.plot(x,y)

plt.legend(['harm. Wellengln.'])
plt.show()
```

```python
%matplotlib inline

ymax = 0.02
TT = 8
lam = 0.2


t = np.arange(0, 50,0.01)

y = ymax * np.sin(2*np.pi * (t/TT - 0.01/lam))

plt.plot(t,y)


#ax.plot_wireframe(x,t,y)
#[Sin[1/0.002*x], {x, 0, 1/360 \[Pi]}]

plt.legend(['t-y Diagramm für  die harm. Wellengln. bei x = 1cm '], loc='lower left')
plt.show()
```
